```python
mean = reviews.points.mean()
result = reviews.points.map(lambda p: p - mean)
pd.concat([result.head(5), result.tail(5)])
```

```python
def remean_points(row):
    row.points = row.points - mean
    return row

reviews.apply(remean_points, axis='columns')
```
